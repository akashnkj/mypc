```python
import pickle
import os



def load_contacts():
    if os.path.exists('phonebook.pkl'):
     with open('phonebook.pkl', 'rb') as file:
      return pickle.load(file)

    else:
       return {}


def save_contacts(contacts):
    with open('phonebook.pkl', 'wb') as file:
     pickle.dump(contacts, file)
    print("Data is saved into a pickle file.")


def main():
    contacts = load_contacts()



    while True:
        print("\nPhonebook Application")
        print("1. Add Contacts")
        print("2. View Contacts")
        print("3. Search Contacts")
        print("4. Update Contacts")
        print("5. Delete Contacts")
        print("6. List Contacts")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_contact(contacts)

        elif choice == "2":
         view_contact(contacts)

        elif choice == "3":
         search_contact(contacts)

        elif choice == "4":
            update_contact(contacts)

        elif choice == "5":
         delete_contact(contacts)

        elif choice == "6":
         list_contact(contacts)

        elif choice == '7':
            break

    else:
       print("Invalid choice. Please enter a number between 1 and 7.")



def add_contact(contacts):
    name = input("Enter new contact name: ")
    if name in contacts:
         print('This contact already exists.')
    else:
        phone = input('Enter contact phone number: ')
        contacts[name] = phone
        print(f"Contact '{name}' added successfully.")
        save_contacts(contacts)

def view_contact(contacts):
   name = input('Enter contact name to view : ')
   if name in contacts:
      print(f"{name}: {contacts[name]}")
   else:
      print("contact not found.")

def search_contact(contacts):
   name = input('Enter the contact you want to search: ')
   if name in contacts :
      print("Search Results: ")
      print(f"{name}: {contacts[name]}")

   else:
        print("No matching contacts found")

def update_contact(contacts):
   name = input('Enter the contact name you want to update: ')
   if name in contacts:
      phone = input('Enter new phone number for this contact name: ')
      if not phone.isdigit():
         print('Invalid phone number. Please enter number in digits.')
         return
      contacts[name] = phone
      save_contacts(contacts)
      print(f"Contact '{name}' updated successfully.")
      print(f"{name}: {contacts[name]}")
   else:
      print("Contact not found.")

def delete_contact(contacts):
   name = input('Enter the contact name to delete: ')
   if name in contacts:
      contact_details = contacts[name]
      del contacts[name]
      save_contacts(contacts)
      print(f"Contact '{name}' with phone number '{contact_details}' deleted successfully")

   else:
      print(f"Contact '{name}' not found. Please check the name and try again.")


def list_contact(contacts):
   if contacts:
      print('list of contacts: ')
      for name, phone in contacts.items():
       print(f"Name: {name} : Phone: {phone}")
      else:
         print("No contacts found.")

if __name__ == '__main__':
   main()
```
